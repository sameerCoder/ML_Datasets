```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_moons
from sklearn.cluster import DBSCAN
```

```python
#data
x,label=make_moons(n_samples=200,noise=0.12)
#data2
#x,label=make_moons(n_samples=200,noise=0)
x[0:5] #independent variable
#basically we dont need dependent variable in unsupervise algo.
```

```python
#label / cluster values
#0 represnt cluster1 and 1 represent cluster2
label
```

```python
# fit the models
#model=DBSCAN(eps=0.25,min_samples=10)#minpts=min_samples
model=DBSCAN(eps=0.25,min_samples=10)
model.fit(x)
```

```python
#model generated cluster value
model.labels_
# -1 means noise
# number of clusters is indicated by model.labels_
```

```python
#Plot
fig,ax=plt.subplots(figsize=(6,5))
ax.scatter(x[:,0],x[:,1],c=model.labels_) # cmap= model.labels_
fig.show()
# 3 colors
```

```python
# with existing correct label dataset.
fig,ax=plt.subplots(figsize=(6,5))
ax.scatter(x[:,0],x[:,1],c=label)
fig.show()
```
