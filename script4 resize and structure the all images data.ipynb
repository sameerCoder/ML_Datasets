```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import PIL.Image as imgmodule 
```

```python
from glob import glob
femalepath=glob('./output_crop/cropfemale/*.png')
print(len(femalepath))
malepath=glob('./output_crop/cropmale/*.png')
allpath=femalepath+malepath
```

```python
# finding one single image size.
print(imgmodule.open(femalepath[0]).size)
#we need either height or width.
print(imgmodule.open(femalepath[0]).size[0])
```

```python
df=pd.DataFrame(allpath,columns=['path'])
df.head()
```

```python
def getsize(path):
        print("current path:",path)
        return imgmodule.open(path).size[0]
```

```python
df['size']=df['path'].apply(getsize)
print(df['size'].head())
```

```python
# removing all images whose size is less than 50
dfnew=df[df['size']>80]
dfnew.head()
```

```python
print(df['path'][0])
print(df['path'][0].split('_')[-2].split('\\')[-1])
```

```python
# create a gender column
def gender(path):
    #print(path)
    try:
        #print("inside try")
        #print(path.split('_')[-2].split('\\')[-1])
        return str(path.split('_')[-2].split('\\')[-1])
    except:
        None
```

```python
dfnew['gender']=dfnew['path'].apply(gender)
dfnew.head()
```

```python
dfnew['gender'].value_counts()
```

```python
dfnew['gender'].value_counts(normalize=True).plot(kind='bar')
plt.show()
```
