```python
import numpy
import cv2 
```

```python
image_read1=cv2.imread("images1.jpg")
print("shape1:",image_read1.shape)
print("Pixel value of image(0 to 255):\n",image_read1)
```

```python
image_read1.shape
# 194 Rows 259 Columns 3 channels- RED, GREEN AND BLUE(RGB).
# OPEN CV READ IN BGR FORMAT- BLUE GREEN RED.
```

```python
# TO DISPLAY THE PICTURE.
from matplotlib import pyplot as plt
plt.imshow(image_read1)
```

```python
# READING image using matplotlib
import matplotlib
matimage=matplotlib.image.imread("images1.jpg")
print(matimage)
plt.imshow(matimage)
# yellowish or same picture as orginal as matplotlib read RED GREEN BLUE(RGB).
```

```python
r=matimage[:,:,0]
g=matimage[:,:,1]
b=matimage[:,:,2]
```

```python
# RGB
plt.figure(figsize=(10,6))
# 1 row 3 columns
plt.subplot(1,3,1)
plt.title('Red Region')
plt.imshow(r,cmap='gray')
plt.subplot(1,3,2)
plt.imshow(g,cmap='gray')
plt.subplot(1,3,3)
plt.title('Blue region')
plt.imshow(b,cmap='gray')
plt.show()
```

```python
# BGR to grayscale # opencv
gray_cv=cv2.cvtColor(image_read1,cv2.COLOR_BGR2GRAY)
plt.imshow(gray_cv,cmap='gray')
```

```python
# grayscale cv image resizing it.
print(gray_cv.shape)
img_re=cv2.resize(gray_cv,(150,200),cv2.INTER_AREA)
print(img_re.shape)
```

```python
plt.imshow(img_re,cmap='gray')
plt.show()
```

```python
# Enlarge(w=2000,h=1600)
img_en=cv2.resize(gray_cv,(2000,1600),cv2.INTER_CUBIC)
img_en.shape
```

```python
plt.imshow(img_en,cmap='gray')
plt.show()
```
