```python
url="https://raw.githubusercontent.com/sameerCoder/ML_Datasets/main/insurance_opt_or_not.csv"
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from tensorflow import keras
import tensorflow as tf
```

```python
df=pd.read_csv(url)
df.head()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df[['Age','Affordability']],df['insurance'],test_size=0.2,random_state=14)
#random state is to fix the random value , same random value will come all time.
```

```python
print(x_train)
```

```python
# As we know machine work better when the value is between 0 &1 , 
# so changing age value between 0 & 1 by dividing by 100.
x_train_scalled=x_train.copy()
x_train_scalled=(x_train_scalled)/100
print(x_train_scalled)

x_test_scalled=x_test.copy()
x_test_scalled=(x_test_scalled)/100
print(x_test_scalled)
```

```python
#https://drive.google.com/file/d/1PfQCBfKiBm2UYcU5L3JQMlb9gDY96iCG/view?usp=sharing
```

![picture](https://drive.google.com/uc?export=view&id=1PfQCBfKiBm2UYcU5L3JQMlb9gDY96iCG)

```python
model1=keras.Sequential([# only 1 layer- WEIGHT 1 & BIAS =0 AS ABOVE PIC.
                         keras.layers.Dense(1,input_shape=(2,),activation='sigmoid',
                                            kernel_initializer='ones',bias_initializer='zeros')
])#log loss and 'binary_crossentropy' are same.
model1.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model1.fit(x_train_scalled,y_train,epochs=6000)
```

```python
model1.evaluate(x_test_scalled,y_test)
```
