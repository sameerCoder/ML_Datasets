```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
#read image
img=cv2.imread('girl1.jpg')
#convert to grayscale.
gray=cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
#apply haar cascade
haar=cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
faces=haar.detectMultiScale(gray,1.3,5)
```

```python
print(faces)
#xcoordinate value, ycoordinate value, width, height.
```

```python
# rectangle - xcoordinate,ycoordinate,xcoordinate+width,ycoordinate+width,boundary thickness
cv2.rectangle(img,(134,43),(134+76,43+76),(0,255,0),3)
plt.imshow(img)
```

```python
# saving only the face crop image.
cv2.imwrite('face1.png',face_crop)
```
